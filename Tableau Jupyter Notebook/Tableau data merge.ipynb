```python
# Parse the date time field into component parts
concat_df['year'] = pd.DatetimeIndex(concat_df['starttime']).year
concat_df['month'] = pd.DatetimeIndex(concat_df['starttime']).month
concat_df['date'] = pd.DatetimeIndex(concat_df['starttime']).date
concat_df['time'] = pd.DatetimeIndex(concat_df['starttime']).hour
concat_df['age'] = concat_df['year'] - concat_df['birth year']
```
